### **Conditional Statements (`if` Structures)**

Conditional statements, or `if` structures, are used in Python to make decisions based on conditions. They enable your code to respond differently depending on whether a condition is `True` or `False`. This is essential for implementing logic in programs, such as filtering data, making classifications, or controlling program flow.

#### **Basic `if` Structure:**

The simplest form of an `if` statement checks a condition, and if that condition is `True`, it executes a block of code:

```python
if condition:
    # Code to execute if the condition is True
```

You can expand this with `elif` (else if) and `else` to check additional conditions or provide a default action:

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition2 is True
else:
    # Code to execute if none of the above conditions are True
```

**Simple Example:**

Here's a basic example where we use an `if` structure to determine if a number is positive, negative, or zero:

```python
# Example: Checking if a number is positive, negative, or zero
number = 5

if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# Output: The number is positive.
```

### **Loops in Python**

Loops are fundamental constructs in Python that allow you to execute a block of code multiple times. They are essential for tasks that involve repetitive actions, such as iterating over data sets, performing calculations on arrays, or processing items in a list. Python offers two main types of loops: `for` loops (counter-controlled loops) and `while` loops (condition-controlled loops).

#### **`for` Loops (Counter-Controlled Loops)**

A `for` loop, also known as a counter-controlled loop, is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each element in the sequence. It’s particularly useful when you know the number of iterations in advance.

**Syntax:**
```python
for element in sequence:
    # Code to execute for each element
```

**Example:**
```python
# Example: Iterating over a list of numbers
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print("Current number:", num)

# Output:
# Current number: 1
# Current number: 2
# Current number: 3
# Current number: 4
# Current number: 5
```

#### **`while` Loops (Condition-Controlled Loops)**

A `while` loop, also known as a condition-controlled loop, repeatedly executes a block of code as long as a specified condition is `True`. It is useful when the number of iterations is not known beforehand and depends on the condition being met.

**Syntax:**
```python
while condition:
    # Code to execute as long as condition is True
```

**Example:**
```python
# Example: Using a while loop to sum numbers until the total exceeds a limit
total = 0
num = 1

while total < 10:
    total += num
    num += 1
    print("Current total:", total)

# Output:
# Current total: 1
# Current total: 3
# Current total: 6
# Current total: 10
```

This distinction helps clarify the different use cases and behaviors of `for` and `while` loops in Python, making it easier to choose the right loop for the task at hand.

Here's the brief warning with an emoji:

⚠️ **Warning:** Be careful with `while` loops! They can produce infinite loops if the evaluated condition is not modified correctly within the loop. Always ensure that the condition will eventually become `False` to avoid your program running indefinitely.

### **Practical Example: Data Processing with Loops**

Loops are often used in data processing tasks, such as calculating statistics, transforming data, or applying operations across multiple data points. Below, we'll use both `for` and `while` loops to compute basic statistics like the mean and standard deviation of a dataset.

#### **Equations:**
- **Mean (Average - $\mu$):** The mean of a set of numbers is calculated as:

  \begin{align}
  \mu &= \frac{1}{n} \sum_{i=1}^{n} x_i
  \end{align}

  where $x_i$ are the data points and $n$ is the total number of data points.

  - **Standard Deviation (std - $\sigma$):** The standard deviation measures the dispersion of data points around the mean, and is calculated as:

  \begin{align}
  \sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
  \end{align}


````python
# Practical example that computes mean and std
data = [1.06, 0.4, -0.90, 0.24, 0.07, -0.98, 0.95, -0.25, -0.41, 0.41]

# Step 1: Calculate the mean using a for loop
sum_data = 0
for num in data:
    sum_data += num

mean = sum_data / len(data)  # Mean calculation

# Step 2: Calculate the standard deviation using a for loop
sum_squared_diff = 0
for num in data:
    sum_squared_diff += (num - mean) ** 2

std_dev = (sum_squared_diff / len(data)) ** 0.5  # Standard Deviation calculation

# Printing the results with .2f precision
print(f"Mean: {mean:.3f}, Standard Deviation: {std_dev:.3f}")  # Output: Mean: 0.16, Standard Deviation: 0.71
````

### **Functions in Python**

Functions are reusable blocks of code that perform a specific task. They help make your code modular, organized, and easier to maintain, especially when dealing with repetitive tasks in data analysis or other applications. By defining functions, you can encapsulate logic into named blocks that can be executed multiple times with different inputs, enhancing code reusability and reducing redundancy.

Using functions aligns with the divide and conquer paradigm, a powerful problem-solving approach where a complex problem is broken down into smaller, more manageable sub-problems. Each sub-problem is solved individually, and their solutions are combined to solve the overall problem. Functions facilitate this paradigm by allowing you to separate your code into distinct units of work, each handling a specific aspect of the problem. This not only simplifies the overall structure of your code but also makes it easier to test, debug, and extend.

#### **Defining Functions**

In Python, functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. The code block within the function is indented, and it executes when the function is called. Functions can take parameters (inputs) and can return a value (output) using the `return` statement.

**Basic Syntax:**

```python
def function_name(parameters):
    # Code block to execute
    return value  # Optional: Return a result
```

- **`def` keyword:** This keyword is used to define a function.
- **Function name:** A descriptive name that follows Python's naming conventions (e.g., snake_case).
- **Parameters:** Variables listed inside the parentheses that accept inputs when the function is called. Functions can have zero or more parameters.
- **Return value:** The result that the function sends back to the caller. Not all functions need to return a value; some may perform actions without returning anything.

#### **Benefits of Using Functions**

1. **Modularity:** Functions allow you to break down complex problems into smaller, manageable pieces, making your code easier to understand and debug.
2. **Reusability:** Once defined, functions can be reused throughout your code without rewriting the same logic, saving time and reducing errors.
3. **Maintainability:** Functions make it easier to update and maintain your code. Changes made inside a function will automatically apply wherever the function is used.
4. **Abstraction:** Functions help abstract away the details of code implementation, allowing you to focus on the higher-level logic of your program.

#### **Using Functions for Data Analysis**

In data analysis, functions are invaluable for tasks like data cleaning, feature extraction, statistical calculations, and model evaluations. By encapsulating these repetitive tasks into functions, you create a toolkit that can be applied consistently across different datasets and analyses.

For example, defining a function to calculate the mean of a dataset allows you to use that function whenever you need to compute the mean, with just a single line of code. Similarly, functions can be used to automate complex workflows, making your data analysis process more efficient and less error-prone.

Functions play a critical role in writing clean, efficient, and professional code, especially in fields like data science, where repetitive tasks are common.

````python
# Sample function
def classify_animal(weight, height):
    """
    Classifies an animal as a 'Cat', 'Dog', or 'Unknown' based on its weight and height.

    Parameters:
    weight (float): The weight of the animal in kilograms.
    height (float): The height of the animal in centimeters.

    Returns:
    str: A classification of the animal ('Cat', 'Dog', or 'Unknown').
    """
    if weight < 10 and height < 30:
        return "Cat"
    elif weight >= 10 and height >= 30:
        return "Dog"
    else:
        return "Unknown"
````

````python
# Example usage of the function with multiple samples
animals = [
    (3, 20),    # Cat sample
    (8, 25),    # Cat sample
    (7, 15),    # Cat sample
    (9, 28),    # Cat sample
    (15, 40),   # Dog sample
    (12, 35),   # Dog sample
    (100, 45),   # Dog sample
    (20, 50),   # Dog sample
]

# List to store classifications
class_animals = []

# Classifying each animal and storing the result
for index, (weight, height) in enumerate(animals, start=1):
    classification = classify_animal(weight, height)
    class_animals.append(classification)
    print(f"Animal {index} is classified as: {classification}")
````

### **Brief Introduction to NumPy**

NumPy (Numerical Python) is a powerful Python library widely used for numerical and scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays. NumPy is fundamental in data science, machine learning, and life sciences because it offers a highly efficient way to perform numerical computations, making data manipulation and analysis faster and more straightforward.

#### **Why NumPy is Efficient**

1. **Performance:** NumPy is written in C, which makes operations on arrays and matrices much faster compared to native Python lists.
2. **Memory Efficiency:** NumPy arrays are stored in contiguous memory locations, which allows efficient access and manipulation of data.
3. **Vectorization:** NumPy allows vectorized operations, which means you can perform operations on entire arrays without writing explicit loops, making the code cleaner and faster.
4. **Broadcasting:** This feature allows NumPy to handle arithmetic operations between arrays of different shapes in a consistent way, which simplifies coding and enhances performance.

#### **Creating Arrays in NumPy**

NumPy arrays (`ndarray`) are the core component of the library. You can create arrays in various ways, such as from Python lists, using built-in functions, or by reading data from files.

**Basic Syntax to Create an Array:**
```python

# Creating a 1D array from a list
array_1d = np.array([1, 2, 3, 4, 5])

# Creating a 2D array (matrix)
array_2d = np.array([[1, 2], [3, 4], [5, 6]])

# Creating an array of zeros
zeros_array = np.zeros((3, 3))  # 3x3 matrix of zeros

# Creating an array of ones
ones_array = np.ones((2, 4))  # 2x4 matrix of ones

# Creating an array with a range of values
range_array = np.arange(0, 10, 2)  # From 0 to 10 with step 2

print("1D Array:", array_1d)
print("2D Array:\n", array_2d)
print("Zeros Array:\n", zeros_array)
print("Ones Array:\n", ones_array)
print("Range Array:", range_array)
```

NumPy supports a variety of operations that can be applied directly to arrays, such as element-wise arithmetic, statistical calculations, and matrix manipulations.

**Examples of Basic Operations:**

In this exercise, we will explore the properties of the normal distribution by generating random data and verifying the empirical rule, also known as the 68-95-99.7 rule.

**Objective:**
We aim to check how many samples fall within the range of $\mu \pm 2\sigma$ and compare the observed percentage with the theoretical expectation of 95.4%.

**Steps:**

4. Compare the results with the theoretical expected value to see if they align with the properties of the normal distribution.

This exercise will help you understand the practical application of the empirical rule in real data, reinforcing the theoretical expectations of normal distributions. Let's see how well our data matches up with the theory!

````python
import numpy as np

# Generating a random vector of 100 elements with mean 0 and standard deviation 1
# np.random.seed(0)  # Setting seed for reproducibility
x = np.random.normal(0, 1, 100)  # mean=0, std=1, size=100

# Calculating the number of elements, mean, and standard deviation using numpy
num_elements = x.size
mean_x = np.mean(x)
std_x = np.std(x)

# Printing the results
print(f"Number of elements: {num_elements}")
print(f"Mean of the data: {mean_x:.2f}")
print(f"Standard Deviation of the data: {std_x:.2f}")

# Testing how many samples fall within the range mean ± 2*std
Ind = (x > mean_x - 2*std_x) & (x < mean_x + 2*std_x)
num_within_range = np.sum(Ind)

# Percent of samples within the range
percent_within_range = num_within_range / num_elements * 100

# Theoretical expected value (approximately 95.4% for a normal distribution within ±2σ)
theoretical_expected = 95.4

print(f"Percent of samples within μ ± 2*σ: {percent_within_range:.2f}%")
print(f"Theoretical expected percent within μ ± 2*σ: {theoretical_expected:.2f}%")
````

Matplotlib and Seaborn are two of the most popular libraries in Python for data visualization. They are powerful tools that help you create a wide variety of plots, allowing you to explore and understand your data better through visual representation.

#### **Matplotlib**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides a lot of flexibility and control over plot appearance, making it a great tool for creating detailed, customized visualizations.

**Key Features of Matplotlib:**
- Highly customizable plots with fine control over every aspect of the plot.
- Supports a wide range of plot types, such as line plots, scatter plots, histograms, bar charts, pie charts, and more.
- Works well with Pandas for quick plotting directly from DataFrames.

**Basic Plotting Functions in Matplotlib:**
- **Scatter Plot:** Used to explore the relationship between two continuous variables.
  ```python

  plt.title("Scatter Plot")

  plt.ylabel("Y-axis")
  ```

Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn makes it easier to create complex visualizations with fewer lines of code and comes with built-in themes for aesthetics.

**Key Features of Seaborn:**
- Simplifies the creation of complex visualizations like violin plots, pair plots, and heatmaps.

#### **Differences Between Matplotlib and Seaborn**

- **Plot Types:** Seaborn includes many specialized plots not available in Matplotlib, such as violin plots and pair plots.
- **Integration:** Seaborn integrates well with Pandas and automatically handles DataFrame inputs, whereas Matplotlib requires more manual handling.

Both libraries are essential tools in exploratory data analysis (EDA) for visualizing data distributions, relationships, and patterns, making them invaluable for gaining insights into datasets.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuring the environment for high-quality figures using Seaborn's style
sns.set_theme(style="whitegrid")  # Set the Seaborn style to whitegrid for clean plots

# Additional Matplotlib configurations for figure size, resolution, and font
plt.rcParams['figure.figsize'] = [12, 8]  # Set the figure size
plt.rcParams['figure.dpi'] = 300          # Set the figure resolution
plt.rcParams['font.family'] = 'sans-serif'  # Use sans-serif fonts for all text
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'Liberation Sans']  # Specify font options
````

````python
# Defining the groups for Age_Mother and Sex_Baby
groups = [(1, 1), (1, 2), (2, 1), (2, 2)]

# Dictionary to store DataFrames for each group
grouped_dfs = {}

# Separating DataFrames using a for loop
for age, sex in groups:
    group_name = f"Age_Mother = {age}, Sex_Baby = {sex}"
    grouped_dfs[group_name] = df[(df['Age_Mother'] == age) & (df['Sex_Baby'] == sex)]

# Creating a 2x2 subplot figure for box plots
fig, axes = plt.subplots(2, 2, figsize=(15, 10),sharey=True)

# Plotting box plots for Wt_Continuous
for i, (group_name, data) in enumerate(grouped_dfs.items()):
    sns.boxplot(y='Wt_Continuous', data=data, ax=axes[i // 2, i % 2])
    axes[i // 2, i % 2].set_title(f'{group_name} - Box Plot of Wt_Continuous')

plt.tight_layout()
plt.show()
````

````python
# Creating a bar plot with means and standard deviations for each group
group_names = []
means = []
stds = []

# Collecting means and standard deviations for each group
for group_name, data in grouped_dfs.items():
    group_names.append(group_name)
    means.append(data['Wt_Continuous'].mean())
    stds.append(data['Wt_Continuous'].std())

# Plotting the bar plot with error bars
plt.figure(figsize=(10, 6))
plt.bar(group_names, means, yerr=stds, capsize=5, color=['skyblue', 'lightgreen', 'salmon', 'orange'])
plt.xlabel('Groups')
plt.ylabel('Mean of Wt_Continuous')
plt.title('Mean and Standard Deviation of Wt_Continuous for Each Group')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
````
